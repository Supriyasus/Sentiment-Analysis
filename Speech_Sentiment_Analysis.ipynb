```python
import matplotlib.pyplot as plt
import string
```

```python
text=open(r"C:\\Users\\supriya shrivastv\\PycharmProjects\\Projects\\Dataset\\speech.txt",encoding="utf-8").read()
```

```python
lower_case=text.lower()
```

```python
lower_case
```

```python
cleaned_text=lower_case.translate(str.maketrans('','',string.punctuation))
cleaned_text
```

```python
#tokenization
tokenised=cleaned_text.split(" ")
tokenised
```

```python
stopwords=["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself",
              "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself",
              "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these",
              "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do",
              "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while",
              "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before",
              "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again",
              "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each",
              "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than",
              "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]

list_of_words=[]
for word in tokenised:
    if word not in stopwords:
        list_of_words.append(word)
list_of_words
```

```python
emotions_list=[]
with open(r"C:\Users\supriya shrivastv\PycharmProjects\Projects\Dataset\emotions.txt","r") as file:
    for line in file:
        clean_line = line.replace("\n","").replace(",","").replace("'","")
        # Ensure there is exactly one colon
        if clean_line.count(':') != 1:
            print("Skipping invalid line: {clean_line}")
            continue
        
        word, emotion = clean_line.split(':')
        word = word.strip()
        emotion = emotion.strip()
        
        if word in list_of_words:
            emotions_list.append(emotion)
```

```python
emotions_list
```

```python
from collections import Counter
count=Counter(emotions_list)
count
```

```python
fig, ax = plt.subplots()
ax.bar(count.keys(), count.values())
fig.autofmt_xdate()  
plt.plot()
```
